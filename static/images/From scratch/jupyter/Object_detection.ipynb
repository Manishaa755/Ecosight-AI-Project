```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
# Define an empty list for class labels
classLabels = []

# Specify the file path
file_name = r"F:\From scratch\label.txt"

# Open the file and read the labels
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')  # Read the file content and split into a list
```

```python
model.setInputSize(320,320)
model.setInputScale (1.0/127.5) # # 255/2 = 127.5
model.setInputMean ((127.5,127.5,127.5)) ##
model.setInputSwapRB (True)
```

### Reading image

```python
img = cv2.imread("F:\From scratch\images\mouse.jpg")
```

```python
plt.imshow(img)
```

```python
# Assuming you have already defined your model and have an image (img) ready for detection
classIndex, confidence, bbox = model.detect(img, confThreshold=0.5)  # Correct the syntax here
```

```python
print(classIndex)
```

```python
# Assuming img, classLabels, ClassIndex, confidence, and bbox are defined correctly

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

# Loop through detected class indices, confidences, and bounding boxes
for classIndex, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
    # Draw the bounding box on the image
    cv2.rectangle(img, boxes, (255, 0, 0), 2)  # Red rectangle

    # Put the class label on the image
    cv2.putText(img, classLabels[classIndex - 1], 
                (boxes[0] + 10, boxes[1] + 40), 
                font, fontScale=font_scale, color=(0, 255, 0), thickness=3)  # Green text
```

```python
import cv2

# Open the webcam (0 is usually the default camera)
cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")

# Font settings
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()  # Read a frame from the webcam
    if not ret:  # Check if the frame was read successfully
        break

    # Object detection
    classIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)  # Use your trained model here

    print(classIndex)  # Print detected class indices for debugging
    if len(classIndex) != 0:
        for classInd, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
            if classInd <= 80:  # Change this based on your classes
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)  # Draw the bounding box
                cv2.putText(frame, classLabels[classInd - 1], (boxes[0] + 10, boxes[1] + 40), 
                            font, fontScale=font_scale, color=(0, 255, 0), thickness=3)  # Add label

    cv2.imshow('Object Detection Tutorial', frame)  # Display the frame with detections
    
    if cv2.waitKey(2) & 0xFF == ord('q'):  # Press 'q' to exit
        break

# Release the webcam and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
